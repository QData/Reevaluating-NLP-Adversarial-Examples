```python
import pandas as pd
mr_df = pd.read_csv('mr-tf-adjusted-1000-successful-examples.csv')
mr_df.head()
```

```python
mr_df[mr_df['label_1'] == mr_df['label_2']]
```

```python
mr_df['plen'] = mr_df['passage_1'].map(len)
mr_df.sort_values(by='plen', inplace=True)
```

```python
pd.set_option('max_colwidth', None)

mr_df['original'] = mr_df['passage_1'].apply(lambda s: s.replace('[[','\color{red}{').replace(']]','}'))
mr_df['perturbed'] = mr_df['passage_2'].apply(lambda s: s.replace('[[','\color{blue}{').replace(']]','}'))
good_mr_df = mr_df[mr_df['label_1'] != mr_df['label_2']]
print('retained', len(good_mr_df), '/', len(mr_df))
sample_mr_df = good_mr_df.sample(n=10)
print('sampled', len(sample_mr_df), '/', len(good_mr_df))
lt = sample_mr_df[['original', 'perturbed']].to_latex(index=False, escape=False)
while ('\n\n' in lt) or ('  ' in lt):
    lt = lt.replace('\n\n','\n').replace('  ', ' ')
print(lt)
```
