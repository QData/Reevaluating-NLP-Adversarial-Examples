```python
import pandas as pd
```

```python
bert_df = pd.read_csv('bert_100_sampled_per_threshold.csv')
use_df = pd.read_csv('use_100_sampled_per_threshold.csv')
```

```python
bert_df.drop('index', axis=1, inplace=True)
bert_df.head()
```

```python
use_df.drop('Unnamed: 0', axis=1, inplace=True)
use_df['SE_Model'] = 'USE'
use_df.head()
```

```python
import csv

full_df = pd.concat((bert_df, use_df))
full_df.rename(columns={ 'passage_1': 'text1', 'passage_2': 'text2'}, inplace=True)
full_df.to_csv('full_100_sampled_per_threshold.csv', index=False, quoting=csv.QUOTE_NONNUMERIC)
```
